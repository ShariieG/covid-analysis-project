```python
import json
#PPLEASE UPLOAD THE FILE ON THE HOME SCREEN BEFORE ATTEMPTING TO RUN THIS CODE
with open('covid.json', 'r') as f:
    data = json.load(f)
```

```python
#THESE ARE THE IMPORTS I AM GOING TO NEED
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#THE JSON FILE CONTAINS DICTIONARIES AND ARRAYS WHICH NEEDS TO BE CONVERTED INTO A DATAFRAME BEFORE BEING USED
dataframe = pd.DataFrame(data)

#I DECIDED TO SORT THE DATA ACCORDING TO THE DATE COLOUMN TO ENSURE DATA INTEGRITY AND ALSO MAKE SURE THE LINE GRAPH MAKES
dataframe= dataframe.sort_values('Date')

#I CONVERTED EACH COLOUMN INTO THE CORRECT DATATYPES, WHICH WILL BE PLOTABLE ON MY GRAPH
dataframe['Date'] = pd.to_datetime(dataframe['Date'])
dataframe['Total Confirmed Cases'] =pd.to_numeric(dataframe['Total Confirmed Cases'], downcast='integer', errors ='coerce')
dataframe['Total Deaths'] =pd.to_numeric(dataframe['Total Deaths'], downcast='integer', errors ='coerce')
dataframe['Total Recovered'] =pd.to_numeric(dataframe['Total Recovered'], downcast='integer', errors ='coerce')
dataframe['Active Cases'] =pd.to_numeric(dataframe['Active Cases'], downcast='integer', errors ='coerce')
dataframe['Daily Confirmed Cases']= pd.to_numeric(dataframe['Daily Confirmed Cases'], downcast='integer', errors ='coerce')
dataframe['Daily  deaths'] =pd.to_numeric(dataframe['Daily  deaths'], downcast='integer', errors ='coerce')
```

```python
#SET THE FIGURE SIZE
plt.figure(figsize = (8, 5))

# I STARTED PLOTING THE DATA/EACH COLOUMN INTO MY GRAPH, GIVING IT LABELS AND COLORS
plt.plot(dataframe['Date'], dataframe['Total Confirmed Cases'], label='Total Confirmed Cases', color='red')
plt.plot(dataframe['Date'], dataframe['Total Deaths'], label='Total Deaths', color='green')
plt.plot(dataframe['Date'], dataframe['Total Recovered'], label='Total Recovered', color='magenta')
plt.plot(dataframe['Date'], dataframe['Active Cases'], label='Active Cases', color='blue')
plt.plot(dataframe['Date'], dataframe['Daily Confirmed Cases'], label='Daily Confirmed Cases', color='black')
plt.plot(dataframe['Date'], dataframe['Daily  deaths'], label='Daily Deaths', color='cyan')

# I ADDED LABELS AND TITLES SO THE GRAPH CAN UNDERSTANDABLE
plt.xlabel('Date')
plt.ylabel('Cases')
plt.title('COVID-19 Data Over Time')

# I DISPLAYED THE LEGEND
plt.legend()

# I SHOWED PLOTS 
plt.show()
```
